```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("zomato.csv")
df.head()
```

```python
df.shape
df.info()
df.isnull().sum()
```

```python
df=df.dropna(subset=['rate (out of 5)','avg cost (two people)'])
df['avg cost (two people)'] = df['avg cost (two people)'].astype(str).str.replace(',','')
df['avg cost (two people)'] = df['avg cost (two people)'].astype(float)
```

```python
plt.scatter(df['avg cost (two people)'],df['rate (out of 5)'])
plt.xlabel('Cost for Two')
plt.ylabel('Rating')
plt.title('Rating vs Cost')
plt.show()
```

```python
df['cuisines type'].value_counts().head(10).plot(kind='bar')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.title('Top 10 Cuisines')
plt.show()
```

```python
df.groupby('online_order')['rate (out of 5)'].mean().plot(kind='bar')
plt.title('Online Order vs Average Rating')
plt.show()
```

##Conclusion
->Expensive restaurants are not always rated higher.  
->Few cuisines dominate the restaurant market.  
->Online ordering restaurants perform a little better.
